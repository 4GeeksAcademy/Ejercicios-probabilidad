## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import numpy as np

cuenta = 0 #variable para llevar la suma de los eventos que buscamos

repeticiones = 1000 #añado esta variable por si se quiere cambiar el numero de repeticiones no tener que estar cambiando el 1000 en cada parte

for i in range(0,repeticiones):
    dado1 = np.random.randint(1,6)
    dado2 = np.random.randint(1,6)
    dado_total = dado1 + dado2
    if (dado_total%2==0) or (dado_total>7): cuenta+=1
    
resultado = cuenta/repeticiones

print("La probabilidad del experimento es de {} ({}%)".format(resultado,(resultado*100)))
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)

primer_caso= 0 #variable para contar cuantas veces se toman 3 bolas blancas y 2 rojas
segundo_caso=0 #variable para contar cuantas veces salen todas las bolas del mismo color           

repeticiones = 1000

for i in range(0,repeticiones):
    #inicio de variables
    blancas = 0
    rojas = 0

    #array de 5 posiciones donde cada posición representa una bola tomada
    bolas = np.random.randint(0,59, size=5)

    #array que transforma los números obtenidos en el color que son
    colores_bolas = [ball_box[i] for i in bolas]
    
    #Bucle para encontrar cuantas veces se toman 3 bolas blancas y 2 rojas
    for color in colores_bolas:
        if color == 'White': blancas+=1
        if color == 'Red': rojas+=1
        if (blancas==3)and (rojas==2): primer_caso+=1
    
    #Bucle para determinar si las bolas que han salido son del mismo color
    if all(color == colores_bolas[0] for color in colores_bolas): segundo_caso+=1

print("\nLa probabilidad de que salgan 3 blancas y 2 rojas es de: {} ({}%)".format((primer_caso/repeticiones),(primer_caso/10)))
print("\nLa probabilidad de que salgan del mismo color es de: {} ({}%)".format((segundo_caso/repeticiones),(segundo_caso/10)))
```
